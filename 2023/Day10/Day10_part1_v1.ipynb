```python
from collections import deque
fileName = 'day10.txt'
#fileName = 'day10_sample.txt'

grid = open(fileName).read().strip().splitlines()


##print(grid)
for r, row in enumerate(grid):
    for c,ch in enumerate(row):
        if ch == 'S':
            sr = r
            sc = c
            break
    else:
        continue
    break

print(sr,sc)
```

```python
seen = {(sr,sc)}
q = deque([(sr,sc)])
#print(seen)
#print(q)

while q:
    r,c = q.popleft()
    ch = grid[r][c]
    #print(r,',',c,' = ',ch)

    if r>0 and ch in 'S|JL' and grid[r-1][c] in '|7F' and (r-1,c) not in seen:
        q.append((r-1,c))
        seen.add((r-1,c))

    if r<len(grid) and ch in 'S|7F' and grid[r+1][c] in '|LJ' and (r+1,c) not in seen:
        q.append((r+1,c))
        seen.add((r+1,c))

    if c>0 and ch in 'S-J7' and grid[r][c-1] in '-LF' and (r,c-1) not in seen:
        q.append((r,c-1))
        seen.add((r,c-1))

    if c<len(grid[r]) and ch in 'S-LF' and grid[r][c+1] in '-J7' and (r,c+1) not in seen:
        q.append((r,c+1))
        seen.add((r,c+1))

print(seen)
print(len(seen) // 2)
print(q)
```
