```python
import re

parsedVales = []
fileName = 'day06.txt'
#fileName = 'day06_sample.txt'

# Use regular expressions to find and extract the values
with open(fileName, 'r') as file:
    for line in file:
        words = line.strip().split()
        #print (words)
        values = words[1:]
        #print(values)
        parsedVales.append(list(map(int,values)))

print(parsedVales)

races = []
for i in range(len(parsedVales[0])):
    races.append((parsedVales[0][i],parsedVales[1][i]))

#races = [(7, 9)]
print(races)
```

```python
noWaysToBeatRecord = 1
for race in races:
    time, distance = race
    couldBeatRecords = []
    print('Time : ',time)
    for t in range(time):
        #print('Sec : ',t)
        #print((time-t)*t)
        d = (time-t)*t
        if(d > distance):
            print('Sec : ',t,' ',d)
            couldBeatRecords.append(t)
    print(couldBeatRecords)
    print(len(couldBeatRecords))
    noWaysToBeatRecord *= len(couldBeatRecords)

print(noWaysToBeatRecord)

    
```
