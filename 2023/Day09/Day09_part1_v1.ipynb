```python
import re

oasisHistory = []
fileName = 'day09.txt'
#fileName = 'day09_sample.txt'

# Use regular expressions to find and extract the values
with open(fileName, 'r') as file:
    for lines in file:
        line = lines.strip().split()
        #print (line)
        line = [int(item) for item in line]
        oasisHistory.append(line)

print(oasisHistory)

```

```python
history_ = [1, 3, 6, 10, 15, 21]

def generateSequence(history):
    seq = []
    for i in range(len(history)-1):
        diff = history[i+1]-history[i]
        #print(diff)
        seq.append(diff)
    return seq

def isSequenceZeros(sequence):
    return all(seq == 0 for seq in sequence)

#print(generateSequence(history_))

def generateSequenceUntilZeros(history):
    historySequences = []
    newSeq = history
    historySequences.append(newSeq)
    while(not isSequenceZeros(newSeq)):
        newSeq = generateSequence(newSeq)
        historySequences.append(newSeq)
    #print(newSeq)
    return historySequences

print(generateSequenceUntilZeros(history_))

```

```python
historySequences = [[1, 3, 6, 10, 15, 21], [2, 3, 4, 5, 6], [1, 1, 1, 1], [0, 0, 0]]

def extrapolateSeq(historySequences):
    historySequences.reverse()
    historySequences[0].append(0)
    seqLength = len(historySequences)

    for i in range(seqLength-1):
        curSeq = historySequences[i]
        nextSeq = historySequences[i+1]
        #print(curSeq,nextSeq)
        nextValue = curSeq[-1] + nextSeq[-1]
        #print(nextValue)
        #historySequences[i+1].append()
        nextSeq.append(nextValue)

    historySequences.reverse()
    #print(historySequences)
    return historySequences[0][-1]

print(extrapolateSeq(historySequences))
```

```python
nextValues = []
for history in oasisHistory:
    generatedHistory = generateSequenceUntilZeros(history)
    print(generatedHistory)
    nextValues.append(extrapolateSeq(generatedHistory))

print(nextValues)

sumOfValues = 0

for value in nextValues:
    sumOfValues += value

print(sumOfValues)
```
