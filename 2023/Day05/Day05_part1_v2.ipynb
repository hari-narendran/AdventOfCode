```python


#fileName = 'day05_sample.txt'     
fileName = 'day05.txt'
seeds, *blocks = open(fileName).read().split('\n\n')

seeds = list(map(int,seeds.split(':')[1].split()))
print(seeds)

for block in blocks:
    ranges = []
    for line in block.splitlines()[1:]:
        ranges.append(list(map(int,line.split())))
    #print(ranges)
    new = []
    for x in seeds:
        for a,b,c in ranges:
            if x in range(b,b+c):
                print(a,' in range of ',b,' + ',c)
                new.append(x-b+a)
                break
        else:
            new.append(x)
    seeds = new

print(seeds)
print(min(seeds))

```
