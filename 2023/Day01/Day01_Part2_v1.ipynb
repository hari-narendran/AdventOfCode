```python
# Open the text file in read mode ('r')
lines=[]
with open('day01.txt', 'r') as file:
    for line in file:
        lines.append(line.strip())

#print(lines)
digitsMap_ = [('one','1'),('two','2'),('three','3'),('four','4'),('five','5'),('six','6'),('seven','7'),('eight','8'),('nine','9')]
digitsMap = [('one','o1e'),('two','t2o'),('three','t3e'),('four','f4r'),('five','f5e'),('six','s6x'),('seven','s7n'),('eight','e8t'),('nine','n9e')]
```

```python
def convertStringToDigit(line):
    #print(line)
    convertedLine=line
    for digit in digitsMap:
        if(digit[0] in line):
            print(digit[0])
            convertedLine = convertedLine.replace(digit[0],digit[1])
            print(convertedLine)
    #print(convertedLine)
    return convertedLine
            
convertStringToDigit('xtwone3four')
        
```

```python
convertedLines = []
for line in lines:
    convertedLines.append(convertStringToDigit(line))
#print(convertedLines)
```

```python
def findCalibrationValue(value):
    digits = []
    for c in value:
        if(c.isdigit()):
            digits.append(c)
    caliValue = ''
    length = len(digits)
    if(length>1):
        caliValue = digits[0]+digits[length-1]
    else:
        caliValue = digits[0]+digits[0]
    return int(caliValue)
```

```python
calibrationValueSum = 0
for line in convertedLines:
    #print(findCalibrationValue(line))
    calibrationValueSum += findCalibrationValue(line)
print(calibrationValueSum)
```
