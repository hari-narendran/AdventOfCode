```python
#fileName = 'day12.txt'
fileName = 'day12_sample.txt'

grid = open(fileName).read().strip().splitlines()

conditions = []
for line in grid:
    damage = line.split(' ')[1]
    damage = damage.split(',')
    damage = [int(ch) for ch in damage]
    obj = {
        'springs': line.split(' ')[0],
        'damage': damage
    }
    conditions.append(obj)

print(grid)
print(conditions)
```

```python

def find_arrangements(springs, damage):
    # Base case: If there are no damaged springs to arrange, return the original 'springs' string.
    if not damage:
        return [springs]
    
    arrangements = []
    first_damage = damage[0]
    
    # Recursively find arrangements for the remaining damaged springs.
    for i in range(len(springs) - first_damage + 1):
        # Try placing damaged springs at position i.
        new_springs = springs[:i] + '#' * first_damage + springs[i+first_damage:]
        
        # Recursively find arrangements for the rest of the damaged springs.
        rest_arrangements = find_arrangements(new_springs, damage[1:])
        
        # Append each arrangement to the list.
        arrangements.extend(rest_arrangements)
    
    return arrangements
row = {'springs': '.??..??...?##.', 'damage': [1, 1, 3]}
arrangements = find_arrangements(row['springs'], row['damage'])
print(arrangements)

```
