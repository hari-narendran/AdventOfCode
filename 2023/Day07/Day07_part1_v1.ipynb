```python
import re

hands = []
fileName = 'day07.txt'
#fileName = 'day07_sample.txt'

# Use regular expressions to find and extract the values
with open(fileName, 'r') as file:
    for line in file:
        words = line.strip().split()
        #print (words)
        cardObj = {
            'hand': words[0],
            'bid': words[1],
            'typeStrength': None,
            'rank': None
        }
        hands.append(cardObj)
        #values = words[1:]
        #print(values)
        #parsedVales.append(list(map(int,values)))

print(hands)


```

```python
def findType(hand):
    #print(hand)
    type = 'High card'
    typeStrength = 1
    counts = []
    for i,card in enumerate(hand):
        #print(card,i,' - ',hand.count(card))
        counts.append(str(hand.count(card)))
    #print(counts)
    for i,card in enumerate(counts):
        #print(card,i,' - ',counts.count(card))
        if(counts.count('5')):
            type = 'Five of a kind'
            typeStrength = 7
        if(counts.count('4')):
            type = 'Four of a kind'
            typeStrength = 6
        if(counts.count('3') == 3 and counts.count('2') == 2):
            type = 'Full House'
            typeStrength = 5
        if(counts.count('3') == 3 and counts.count('1') == 2):
            type = 'Three of a kind'
            typeStrength = 4
        if(counts.count('2') == 4 and counts.count('1') == 1):
            type = 'Two pair'
            typeStrength = 3
        if(counts.count('2') == 2 and counts.count('1') == 3):
            type = 'One pair'
            typeStrength = 1
    #print(type)
    return typeStrength


print(findType('AAAAA'))
```

```python
for hand in hands:
    hand['typeStrength'] = findType(hand['hand'])
#print(hands)

sortedHands = sorted(hands, key = lambda x: x['typeStrength'])
#print(sortedHands)

groupedHands = {}
for hand in sortedHands:
    typeStrength = hand['typeStrength']
    if typeStrength not in groupedHands:
        groupedHands[typeStrength] = []
    groupedHands[typeStrength].append(hand)

print(groupedHands)
```
