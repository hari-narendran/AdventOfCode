```python
fileName = 'day11.txt'
#fileName = 'day11_sample.txt'

grid = open(fileName).read().strip().splitlines()

#print(grid)

emptyRows = [r for r ,row in enumerate(grid) if '#' not in row]
emptyCols = [c for c, col in enumerate(zip(*grid)) if '#' not in col]

#print(emptyRows)
#print(emptyCols)

points = [(r,c) for r,row in enumerate(grid) for c,ch in enumerate(row) if ch == '#']

print(points)
```

```python
def manhattanDistance(p1,p2):
    x1,y1 = p1
    x2,y2 = p2
    x1,x2 = min(x1,x2), max(x1,x2)
    y1,y2 = min(y1,y2), max(y1,y2)
    dist = 0
    #print(x1,y1)
    #print(x2,y2)

    for x in range(x1,x2):
        #print(x)
        dist += 1
        if x in emptyRows:
            #print('Empty Row')
            dist += 1
    for y in range(y1,y2):
        #print(y)
        dist += 1
        if y in emptyCols:
            #print('Empty Col')
            dist += 1
    #print(' - ',dist)
    return dist

#manhattanDistance((0, 3),(8, 7))
```

```python
pairs = []
sumOfSteps = 0
for i in range(len(points)):
    for j in range(i+1,len(points)):
        #print(points[i],points[j])
        sumOfSteps += manhattanDistance(points[i], points[j])
print(sumOfSteps)
```
